```python
import requests
import pandas as pd

def fetch_api_data(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        data = response.json()
        print("Data fetched from API:")
        return data
    except Exception as e:
        print("API fetch error:", e)
        return None

def compute_average(numbers):
    if not numbers:
        return 0
    return sum(numbers) / len(numbers)

def count_items(items):
    counts = {}
    for item in items:
        counts[item] = counts.get(item, 0) + 1
    return counts

if __name__ == "__main__":
    api_url = "https://jsonplaceholder.typicode.com/posts"
    api_data = fetch_api_data(api_url)

    sample_numbers = [10, 20, 30, 40, 50]
    avg = compute_average(sample_numbers)
    print(f"Average of sample numbers: {avg}")

    sample_items = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']
    item_counts = count_items(sample_items)
    print("Item counts:", item_counts)

    df = pd.DataFrame(list(item_counts.items()), columns=['Item', 'Count'])
    print("\nDataFrame from item counts:")
    print(df)
```
