```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Chart Visualization Examples', fontsize=18, fontweight='bold')

x = np.linspace(0, 10, 11)
y = x ** 2
axes[0, 0].plot(x, y, marker='o', color='#21808d', linewidth=2.5,
                markersize=8, markerfacecolor='#32b8c6', markeredgecolor='#21808d')
axes[0, 0].set_title('X,Y Line Graph (Y = X²)', fontweight='bold', fontsize=14)
axes[0, 0].set_xlabel('X Axis', fontsize=11)
axes[0, 0].set_ylabel('Y Axis', fontsize=11)
axes[0, 0].grid(True, alpha=0.3, linestyle='--')
axes[0, 0].legend(['Y = X²'], loc='upper left', fontsize=10)

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [65, 59, 80, 81, 56, 55]
colors_bar = ['#21808d', '#32b8c6', '#2996a1', '#1d7480', '#1a6873', '#2da6b2']

bars = axes[0, 1].bar(months, sales, color=colors_bar, alpha=0.85,
                       edgecolor='black', linewidth=1.2)
axes[0, 1].set_title('Bar Chart - Monthly Sales', fontweight='bold', fontsize=14)
axes[0, 1].set_xlabel('Months', fontsize=11)
axes[0, 1].set_ylabel('Sales Values', fontsize=11)
axes[0, 1].grid(axis='y', alpha=0.3, linestyle='--')

for bar in bars:
    height = bar.get_height()
    axes[0, 1].text(bar.get_x() + bar.get_width()/2., height,
                    f'{int(height)}', ha='center', va='bottom', fontsize=9)

np.random.seed(42)
data = np.random.normal(35, 15, 200)

axes[1, 0].hist(data, bins=10, color='#21808d', alpha=0.7,
                edgecolor='black', linewidth=1.2)
axes[1, 0].set_title('Histogram - Frequency Distribution', fontweight='bold', fontsize=14)
axes[1, 0].set_xlabel('Data Ranges', fontsize=11)
axes[1, 0].set_ylabel('Frequency', fontsize=11)
axes[1, 0].grid(axis='y', alpha=0.3, linestyle='--')

mean_val = np.mean(data)
axes[1, 0].axvline(mean_val, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_val:.1f}')
axes[1, 0].legend(fontsize=10)

languages = ['Python', 'JavaScript', 'Java', 'C++', 'R']
usage = [30, 25, 20, 15, 10]
colors_pie = ['#21808d', '#32b8c6', '#2996a1', '#1d7480', '#a84b2f']
explode = (0.05, 0, 0, 0, 0)

wedges, texts, autotexts = axes[1, 1].pie(usage, labels=languages, autopct='%1.1f%%',
                                            colors=colors_pie, startangle=90,
                                            explode=explode, shadow=True)
axes[1, 1].set_title('Pie Chart - Programming Language Usage', fontweight='bold', fontsize=14)

for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontsize(10)
    autotext.set_fontweight('bold')

plt.tight_layout()

plt.show()
```
