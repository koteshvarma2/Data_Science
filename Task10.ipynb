```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt

np.random.seed(42)
num_samples = 300

X = pd.DataFrame({
    'feature1': np.random.normal(5, 2, num_samples),
    'feature2': np.random.normal(10, 3, num_samples),
    'feature3': np.random.normal(15, 5, num_samples)
})

y = np.random.choice([0, 1, 2, 3], size=num_samples)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)


model = DecisionTreeClassifier(random_state=42, max_depth=4)
model.fit(X_train, y_train)


y_pred = model.predict(X_test)


accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}\n")
print("Classification Report:")
print(classification_report(y_test, y_pred, target_names=['A', 'B', 'AB', 'O']))


plt.figure(figsize=(12, 6))
plot_tree(
    model,
    feature_names=X.columns,
    class_names=['A', 'B', 'AB', 'O'],
    filled=True,
    rounded=True,
    fontsize=6
)
plt.title("Decision Tree Model for Multi-Class Classification")
plt.show()
```
