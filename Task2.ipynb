```python
import pandas as pd

data = {
    'Age': [23, 45, 31, 35, 58, 40, 38, 31, 29, 41],
    'Salary': [50000, 82000, 60000, 72000, 91000, 68000, 72000, 62000, 58000, 70000]
}

df = pd.DataFrame(data)

description = df.describe()

age_mode = df['Age'].mode()[0]
salary_mode = df['Salary'].mode()[0]

print("Descriptive Statistics:\n", description)
print(f"\nMode of Age: {age_mode}")
print(f"Mode of Salary: {salary_mode}")
```
